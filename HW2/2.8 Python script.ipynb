```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dimensions = range(1, 40+1)
samples_per_dimension = 100
sigma = 1.0       # standard deviation
```

```python
mean_radii = []
std_radii = []

for m in dimensions:
    samples = np.random.normal(loc = 0, scale = sigma, size=(samples_per_dimension, m))
    radii = np.linalg.norm(samples, axis = 1)
    mean_radii.append(np.mean(radii))
    std_radii.append(np.std(radii))


plt.figure(figsize=(10, 6))
plt.plot(dimensions, mean_radii, label='Mean Radius', marker='x', linestyle='-')
plt.plot(dimensions, std_radii, label='Standard Deviation of Radius', marker='x', linestyle='-')
plt.xlabel('Dimension (m)')
plt.ylabel('Radius')
plt.title('Mean and Standard Deviation of Radii of Samples from m-Dimensional Gaussian')
plt.legend()
plt.grid(True)
plt.show()
```

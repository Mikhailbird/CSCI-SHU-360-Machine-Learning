## For problem 3.1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge

# initialize points
np.random.seed(42)
X = np.linspace(0, 10, 10).reshape(-1, 1)  # Feature data
y = 3 * X + 2  # Linear relationship with no noise
plt.scatter(X, y, color='blue', label='Data Points')

# Fit the model
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

ridge_model = Ridge(alpha=10)
ridge_model.fit(X, y)
y_pred_ridge = ridge_model.predict(X)


plt.plot(X, y_pred_linear, color='green', label='Linear Regression', linestyle='--')
plt.plot(X, y_pred_ridge, color='red', label='Ridge Regression', linestyle='--')
plt.title('Standard Linear Regression vs Ridge Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.show()
```

## For Problem 3.2

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge

# Generate random data
np.random.seed(0)
X = 2 * np.random.rand(25, 1)
y = 1 + 2 * X.flatten() + np.random.randn(25)
x_range = np.linspace(min(X), max(X), 100)




X = np.append(X, [[0.5], [0.7], [0.8], [1.0], [1.2], [1.7]], axis=0)
y = np.append(y, [10, 11, 8, 9, 15, 16])


X = X.reshape(-1, 1)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(x_range, 1 + 2.2 * x_range, color='red', linewidth=2, label = 'ridge regression')
plt.plot(x_range, 0.8 + 4 * x_range, color='green', linewidth=2, label = 'linear regression')

plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('linear regression vs ridge regression')
plt.show()
```
